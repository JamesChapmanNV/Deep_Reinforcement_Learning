```python
import matplotlib.pyplot as plt
import numpy as np

from IPython import display
plt.ion()

np.random.seed(0)
```

## Plotting

```python
def plot_rewards(DATA_SAVE_PATH, bottom, top):
    
    total_rewards_file = 'data/{}_total_rewards.npy'.format(DATA_SAVE_PATH)
    mean_rewards_file = 'data/{}_mean_rewards.npy'.format(DATA_SAVE_PATH)
    std_rewards_file = 'data/{}_std_rewards.npy'.format(DATA_SAVE_PATH)
    epsilons_file = 'data/{}_epsilons.npy'.format(DATA_SAVE_PATH)
    frame_idxs_file = 'data/{}_frame_idxs.npy'.format(DATA_SAVE_PATH)


    with open(total_rewards_file, 'rb') as f:
        total_rewards = np.load(total_rewards_file)

    with open(mean_rewards_file, 'rb') as f:
        mean_rewards = np.load(mean_rewards_file)

    with open(std_rewards_file, 'rb') as f:
        std_rewards = np.load(std_rewards_file)

    with open(epsilons_file, 'rb') as f:
        epsilons = np.load(epsilons_file)

    with open(frame_idxs_file, 'rb') as f:
        frame_idxs = np.load(frame_idxs_file)


    total_rewards = total_rewards.tolist()       
    mean_rewards = mean_rewards.tolist()       
    std_rewards = std_rewards.tolist()       
    epsilons = epsilons.tolist()       
    frame_idxs = frame_idxs.tolist()    
    
    
    fig, ax1 = plt.subplots() 
    ax1.set_xlabel('Frames')
    ax1.set_ylabel('Reward') 
    ax1.plot(frame_idxs,mean_rewards, label = "Mean Training Reward", color='Red')
    ax1.fill_between(frame_idxs, np.array(mean_rewards) - np.array(std_rewards), np.array(mean_rewards) + np.array(std_rewards), color='b', alpha=0.1)
    ax1.legend(loc ="upper right")
    ax1.set_ylim(bottom=bottom,top=top)
    
    ax2 = ax1.twinx() 
    ax2.set_ylabel('epsilons', color = 'Green') 
    ax2.plot(frame_idxs,epsilons, label = "epsilons", color='Green')
    ax2.tick_params(axis ='y', labelcolor = 'Green') 

    plt.title(DATA_SAVE_PATH) 
```

```python
plot_rewards('Breakout-DQN',0,80)
plot_rewards('Breakout-DOUBLE',0,80)
plot_rewards('Breakout-PER',0,80)
plot_rewards('Breakout-DUELING',0,80)

plot_rewards('Pong-DQN',-25,30)
plot_rewards('Pong-DOUBLE',-25,30)
plot_rewards('Pong-PER',-25,30)
plot_rewards('Pong-DUELING',-25,30)
```
